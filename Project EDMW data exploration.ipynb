Import statements

```python
from bs4 import BeautifulSoup
import requests
```

Text extraction

```python
data = requests.get("https://forums.hardwarezone.com.sg/eat-drink-man-woman-16/pm-lee-sinkie-chinese-diff-tiongs-6097816.html").text
soup = BeautifulSoup(data, 'html.parser')
```

```python
print(soup.prettify())
```

```python
post_list = soup.findAll("div", {"class": "post_message"})
posts = [i.string for i in post_list]
```
